## Part D: PCA (Dimensionality Reduction)

```python
hr.info()
```

```python
# Standardize numerical features before applying PCA.
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

drop_cols = ['EmpID','ManagerID','DeptID','MarriedID','MaritalStatusID',
              'GenderID','EmpStatusID','PositionID','PerfScoreID','Zip']
numeric_columns = numeric_columns.drop(columns=drop_cols, errors='ignore')

imputer = SimpleImputer(strategy='mean')
hr_imputed = imputer.fit_transform(numeric_columns)

scaler = StandardScaler()
hr_scaled = scaler.fit_transform(hr_imputed)
```

```python
# Perform PCA on the dataset and explain the first 2 components.
from sklearn.decomposition import PCA

pca = PCA()
pca_data = pca.fit_transform(hr_scaled)

explained_var = pca.explained_variance_ratio_
components = pca.components_

print("Explained Variance Ratio:", explained_var)
print("PCA Components (loadings):")
print(pd.DataFrame(components,
                   columns=numeric_columns.columns,
                   index=[f'PC{i+1}' for i in range(components.shape[0])]))
```

First 2 explain about 36.7% of the variance.

```python
# Plot the explained variance ratio for the PCA components.
plt.figure(figsize=(8,5))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
         pca.explained_variance_ratio_, marker="o")
plt.title("Scree Plot - Explained Variance by PCA Component")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.grid(True)
plt.show()
```

```python
# Reduce the dataset to 2 dimensions with PCA and plot employees colored by department.
pca_2d = PCA(n_components=2)
pca_data = pca_2d.fit_transform(hr_scaled)

# Create DataFrame for plotting
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_df['Department'] = hr['Department']

# Scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Department', alpha=0.7)
plt.title("Employees in PCA Space (colored by Department)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Department')
plt.show()
```

```python
# Which variables contribute most to the first principal component?
loadings = pd.DataFrame(
    pca.components_,          # each row = principal component
    columns=numeric_columns.columns,   # original feature names
    index=[f"PC{i+1}" for i in range(pca.n_components_)]
)
pc1_loadings = loadings.loc["PC1"].sort_values(ascending=False)
plt.figure(figsize=(8,5))
pc1_loadings.plot(kind="bar")
plt.title("Feature Contributions to Principal Component 1")
plt.ylabel("Loading Score")
plt.show()
```

```python
# Apply PCA to engagement + satisfaction + absences. Do these condense into 1 dimension?
# Convert scaled data back into DataFrame
hr_scaled_df = pd.DataFrame(hr_scaled, columns=numeric_columns.columns)

# Select wellbeing features
wellbeing_features = hr_scaled_df[['EngagementSurvey', 'EmpSatisfaction', 'Absences']]

# Apply PCA
pca_wellbeing = PCA()
pca_wellbeing.fit(wellbeing_features)

print("Explained Variance Ratio:", pca_wellbeing.explained_variance_ratio_)
```

```python
# Visualize employees in PCA-reduced 2D space grouped by PerformanceScore.
plt.figure(figsize=(8,6))
sns.scatterplot(
    x=pca_data[:,0], y=pca_data[:,1],
    hue=hr['PerformanceScore'],
    palette='viridis', alpha=0.7
)
plt.title("Employees in PCA Space (colored by Performance Score)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Performance Score", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Compare clustering before and after PCA using k-means on numeric data.
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Clustering on raw numeric data
kmeans_raw = KMeans(n_clusters=3, random_state=42)
labels_raw = kmeans_raw.fit_predict(hr_scaled)
sil_raw = silhouette_score(hr_scaled, labels_raw)

# Clustering on PCA-reduced data (2D or top N PCs)
kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(pca_data[:, :2])  # first 2 PCs
sil_pca = silhouette_score(pca_data[:, :2], labels_pca)

print("Silhouette Score (Raw):", sil_raw)
print("Silhouette Score (PCA 2D):", sil_pca)
```

```python
# Plot loadings of salary, absences, and engagement on the first 2 PCs.
# Create a scatterplot of employees in PC1â€“PC2 space
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:,0], pca_data[:,1], alpha=0.4)

# Get loadings for selected features
features = ['Salary', 'Absences', 'EngagementSurvey']
loadings = pca_2d.components_.T  # shape: (n_features, n_components)
selected_idx = [list(numeric_columns.columns).index(f) for f in features]

# Plot arrows for loadings
for i, idx in enumerate(selected_idx):
    plt.arrow(0, 0,
              loadings[idx,0]*5, loadings[idx,1]*5,  # scale up for visibility
              color='red', alpha=0.8, head_width=0.1)
    plt.text(loadings[idx,0]*5*1.1, loadings[idx,1]*5*1.1,
             features[i], color='red', fontsize=12)

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Biplot: Salary, Absences, Engagement")
plt.grid(True)
plt.show()
```

## Discuss when PCA is appropriate in HR analytics (e.g., reducing survey dimensions).
PCA is most useful when you have many correlated numeric variables and want to reduce dimensionality while preserving as much variance (information) as possible.

In HR analytics, PCA makes sense in cases like:

- Employee surveys: Engagement, satisfaction, culture-fit, motivation, etc. are often correlated. PCA can condense them into fewer factors (e.g., â€œoverall engagementâ€).

- Performance metrics: If employees have multiple scores (productivity, quality, timeliness, collaboration), PCA can reduce them into 1â€“2 overall performance indices.

- Absenteeism & time-tracking: Days late, absence hours, leave counts might be redundant. PCA can summarize them into a single â€œattendanceâ€ dimension.

- Skill assessments: Multiple test scores may be compressed into fewer latent skill dimensions.

When not appropriate:

- If variables are categorical (e.g., department, job role). PCA assumes continuous numeric data.

- If interpretability is more important than dimensionality reduction. (HR managers often prefer simple, clear metrics over abstract PCs.)

- If features are already uncorrelated (then PCA doesnâ€™t add value)
